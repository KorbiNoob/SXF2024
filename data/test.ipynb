```python
import json
from uuid import uuid4
import os
with open("database.json", mode="r") as f:
    dic = json.load(f)
valid_ids = list(map(lambda x: x.split(".")[0],os.listdir("../static/photos/")))
```

```python
dic
```

```python
for entry in dic:
    if entry['photo'] not in valid_ids:
        entry['photo'] = "standard"
```

```python
with open("database.json", mode="w", encoding="utf-8") as f:
    json.dump(obj=dic, fp=f, ensure_ascii=False)
```
